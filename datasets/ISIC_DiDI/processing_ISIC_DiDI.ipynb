```python
import os
CWD = os.getcwd()
import numpy as np
import pandas as pd 
import shutil
import PIL
from PIL import Image
import tensorflow as tf
from tensorflow import image
from tensorflow.image import ResizeMethod
import derm_ita as dita
import h5py

print(CWD)
```

```python
metadata_ISIC_DiDI = pd.read_excel(f'{CWD}/metadata_ISIC_DiDI.xlsx', index_col='index')
print(metadata_ISIC_DiDI)
```

```python
print(os.listdir(f'{CWD}/all_images'))
print(len(os.listdir(f'{CWD}/all_images')))
```

```python
metadata_ISIC_DiDI['id'] = os.listdir(f'{CWD}/all_images')
```

```python
print(metadata_ISIC_DiDI)
num_imgs = len(metadata_ISIC_DiDI['id'])
```

```python
images = np.empty(shape=[num_imgs, 64, 64, 3], dtype=np.uint8)
for i in range(num_imgs):
    id = metadata_ISIC_DiDI['id'][i]
    image = tf.keras.utils.load_img(f'{CWD}/all_images/{id}', target_size = (64, 64))
    image = tf.keras.utils.img_to_array(image, dtype=np.uint8)
    images[i] = image
```

```python
with h5py.File(f'{CWD}/ISIC_DiDI.h5','w') as f:
    f.create_dataset('images', data = images)
    f.create_dataset('labels', data = np.array(metadata_ISIC_DiDI['malignance'], dtype=np.uint8).reshape(num_imgs,1))
```
