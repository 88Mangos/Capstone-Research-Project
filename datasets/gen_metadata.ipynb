```python
import pandas as pd 
import os
CWD = os.getcwd()
filenames = []
labels = []
data = []
F = ["ISIC_2017", "DiDI_2022"]
L = ["benign", "malignant"]
```

```python
for folder in F:
    for label in L:
        contents = os.listdir(f'{CWD}/{folder}/{label}')
        for image in contents:
            datum = []
            if not (len(image)==20):
                print("error") #----_20##_######.---
            filenames.append(image)
            labels.append(label)
            datum.append(image)
            datum.append(label)
            data.append(datum)
```

```python
print(filenames)
print(labels)
print(len(filenames))
print(len(labels))
```

```python
metadata = pd.DataFrame(data = data, columns = ['filename','label'])

print(metadata)
```

```python
metadata.to_csv(f'{CWD}/metadata.csv')
```
